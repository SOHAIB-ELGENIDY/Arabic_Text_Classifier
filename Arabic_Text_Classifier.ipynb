```python
stop_words = stopwords.words()
```

```python
punctuations = '''`÷×؛<>_()*&^%][ـ،/:"؟.,'{}~¦+|!”…“–ـ''' + string.punctuation
```

```python
def clean(text):
    
    RemovePunc = str.maketrans('', '', punctuations)
    text = text.translate(RemovePunc)
    
    #remove longation
    text = re.sub("[إأآا]", "ا", text)
    text = re.sub("ة", "ه", text)
    text = re.sub("ى", "ي", text)
    text = re.sub("ؤ", "ء", text)
    text = re.sub("گ", "ك", text)
    text = re.sub("ئ", "ء", text)
    
    

    text = ' '.join(word for word in text.split() if word not in stop_words)

    return text
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    clean_words = []
    for word in text.split():
        clean_words.append(ps.stem(word))
        
    return ' '.join(clean_words)
```

```python
stem(df.Sentences[3])
```

```python
df['Sentences'] = df['Sentences'].apply(clean).apply(stem)
df.head(3)
```

```python
cv.fit(df.Sentences)
```

```python
x = cv.transform(df.Sentences).toarray()#.shape
```

**Models**

```python
x = df.Sentences
y = df.Sentiment
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size =.10, random_state=90)
```

**LogisticRegression**

```python
p = make_pipeline(TfidfVectorizer(),LogisticRegression())
L_grid = {'logisticregression__C': [0.01, 0.1, 1, 10, 100]}
```

```python
model = GridSearchCV(p, L_grid, cv=5)
model.fit(X_train,Y_train)
print('Done')
```

```python
LR_Prediction = model.predict(X_test)
print(f"Accuracy score is {accuracy_score(Y_test, LR_Prediction):.2f}")
print(classification_report(Y_test, LR_Prediction))
```

**RandomForestClassifier**

```python
p = make_pipeline(TfidfVectorizer(),
                    RandomForestClassifier())

R_grid = {'randomforestclassifier__n_estimators':[10, 100, 1000],
             'randomforestclassifier__max_features':['sqrt', 'log2']}
```

```python
Random = GridSearchCV(p, R_grid, cv=5)
Random.fit(X_train,Y_train)
print('Done')
```

```python
RF_Prediction = Random.predict(X_test)
print(f"Accuracy score is {accuracy_score(Y_test, RF_Prediction):.2f}")
print(classification_report(Y_test, RF_Prediction))
```
